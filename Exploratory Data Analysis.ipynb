```python
fig,axes=plt.subplots(1,5,figsize=(12,6),sharey=True)
sns.kdeplot(ax=axes[0],data=game['NA_Sales'],shade=True)
axes[0].set_title('NA_sales')

sns.kdeplot(ax=axes[1],data=game['EU_Sales'],shade=True)
axes[1].set_title('EU_Sales')

sns.kdeplot(ax=axes[2],data=game['JP_Sales'],shade=True)
axes[2].set_title('JP_Sales')

sns.kdeplot(ax=axes[3],data=game['Other_Sales'],shade=True)
axes[3].set_title('Other_Sales')

sns.kdeplot(ax=axes[4],data=game['Global_Sales'],shade=True)
axes[4].set_title('Global_Sales')
```
